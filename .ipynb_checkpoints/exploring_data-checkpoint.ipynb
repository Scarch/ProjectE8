Importing libraries

```python
import pandas as pd
```

Reading in kaggle data:

```python
kaggle_training_data = pd.read_csv("kaggleData/train.csv")
kaggle_test_data = pd.read_csv("kaggleData/test.csv")
```

```python
kaggle_test_data.describe()
```

Checking if registered and casual user counts add up to the total user count:

```python
mistakes = 0

for index in kaggle_training_data.index:
    casual = kaggle_training_data.casual[index]
    registered = kaggle_training_data.registered[index]
    count = kaggle_training_data["count"][index]

    total = casual + registered
    if total != count:
        mistakes += 1
        print(f"Mistake in row {index}: casual({casual}) + registered({registered}) = {total}, but actual count is {count}")


print(f"Done, mistakes: {mistakes}")
```

Adding training data and test data together for data analysis:

```python
kaggle_data = pd.concat([kaggle_training_data, kaggle_test_data], ignore_index=True).drop(["casual", "registered", "count"], axis=1)
```

Checking if temperature and "feels like" temperature differences are in a normal range:

```python
kaggle_data['temp_atemp_diff'] = abs(kaggle_data['temp'] - kaggle_data['atemp'])
```

```python
kaggle_data.temp_atemp_diff.describe()
```

Found some anomalies:

```python
for index in kaggle_data.index:
    if 20 < kaggle_data.temp_atemp_diff[index]:
        print(f"{index}. row needs to be checked:")
        print(kaggle_data.iloc[index])
        print()
```

```python
windspeed_zero = kaggle_data[kaggle_data['windspeed'] == 0.0]
windspeed_zero.plot.scatter(x='temp', y='atemp', ylabel="feels like temp", alpha=0.5, title='Temp vs "feels like" temp for Windspeed = 0.0')
```

```python
count = 0

for index in kaggle_data.index:
    if 0 == kaggle_data.windspeed[index] and kaggle_data.weather[index] == 4:
        count += 1
        print(f"{index}. row needs to be checked:")
        print(kaggle_data.iloc[index])
        print()

print(f"Done, {count} anomalies found")
```

Checking for anomalies in humidity:

```python
for index in kaggle_data.index:
    if 0 == kaggle_data.humidity[index]:
        print(f"{index}. row needs to be checked:")
        print(kaggle_data.iloc[index])
        print()
```

```python
rows = 0

for index in kaggle_data.index:
    if 100 == kaggle_data.humidity[index] and 20 <= kaggle_data.windspeed[index]:
        rows += 1
        print(f"{index}. row needs to be checked:")
        print(kaggle_data.iloc[index])
        print()

print(rows)
```
